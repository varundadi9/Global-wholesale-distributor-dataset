```python
from pyspark.sql.functions import col

fact_payment = spark.table("03_gold.gold.fact_payment").drop("invoice_id", "invoice_status")

df = spark.table("03_gold.gold.dim_customer") \
    .join(spark.table("03_gold.gold.dim_regions"), on="country") \
    .join(spark.table("03_gold.gold.fact_invoice_line_item"), on="customer_id") \
    .join(spark.table("03_gold.gold.dim_products"), on="product_id") \
    .join(fact_payment, on="customer_id")

display(df)

df.write.format("delta").mode("overwrite").saveAsTable("03_gold.gold.customer_data_cube")
```
